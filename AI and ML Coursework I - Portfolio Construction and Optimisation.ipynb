<span style="color: DarkBlue;">**TO CREATE AN EFFICIENT FRONTIER, EMPTY ARRAYS ARE CREATED FOR THE RETURNS, RISK, AND WEIGHTS.**</span>

<span style="color: DarkBlue;">**AFTERWARDS, A LOOP OF 100,000 PORTFOLIOS IS CREATED, THEN RANDOM ALLOCATION OF WEIGHTS IS CODED TO TRY DIFFERENT COMBINATION OF WEIGHTS BETWEEN THE FUNDS TO EVENTUALLY FIND THE "MINIMUM VOLATILITY PORTFOLIO" AND THE "OPTIMAL RISKY PORTFOLIO".**</span>



```python
p_ret = [] 
p_vol = []
p_weights = []
num_assets = len(data3.columns)
num_portfolios = 100000
```

```python
for portfolio in range(num_portfolios):
    weights = np.random.random(num_assets)
    weights = weights/np.sum(weights)
    p_weights.append(weights)
    returns = np.dot(weights, Annual_Returns3)

    p_ret.append(returns)
    var = Returns_Covariance3.mul(weights, axis=0).mul(weights, axis=1).sum().sum()
    sd = np.sqrt(var) 
    ann_sd = sd*np.sqrt(12)
    p_vol.append(ann_sd)
```

```python
datafinal = {'Returns':p_ret, 'Risk':p_vol}

for counter, symbol in enumerate(data3.columns.tolist()):
    datafinal[symbol+'weight'] = [w3[counter] for w3 in p_weights]
portfolios = pd.DataFrame(datafinal)
portfolios.head()
```

```python
portfolios.plot.scatter(x='Risk', y='Returns', marker='o', s=10, alpha=0.3, grid=True, figsize=[5,5])
```

```python
min_vol_port = portfolios.iloc[portfolios['Risk'].idxmin()]
min_vol_port
```

```python
plt.subplots(figsize=[6,5])
plt.scatter(portfolios['Risk'], portfolios['Returns'],marker='o', s=10, alpha=0.3)
plt.scatter(min_vol_port[1], min_vol_port[0], color='r', marker='*', s=200)
```

```python
rf = 0.035
optimal_risky_port = portfolios.iloc[((portfolios['Returns']-rf )/portfolios['Risk']).idxmax()]
optimal_risky_port
```
